```python
# Import the functions to create the appropriate network data
from create_network import *

# Load and clean the data
wiki = open_file('../data/rowiki_2006.txt')

# Preprocess the data to calculate seniority
username_dates = preprocess_data(wiki)

# Create the network data
wiki_net, num_items = make_network_data(wiki)

# Print the first 5 data points and the number of nodes and edges
print(wiki_net[:5])
print(num_items)
```

```python
# Import the function for counting the number of AB-BA event sequences
from abba import count_ab_ba_events

# Get the number of AB-BA event sequences in the data and store data about each event
num_abba, abba_events = count_ab_ba_events(wiki_net)

# Print the number of AB-BA event sequences
print(num_abba)
```

```python
# Import needed functions
from abba import separate_abba, abs_seniority_diff, plot_histograms, mean_diff

# Separating data into AB-BA edges and non-AB-BA edges
abba_edges, non_abba_edges = separate_abba(wiki_net, abba_events)

# Get absolute seniority difference for reverts in AB-BA sequences and the rest
abba_diff = abs_seniority_diff(abba_edges)
other_diff = abs_seniority_diff(non_abba_edges)

# Plot a histogram with the two distributions overlapping
plot_histograms(abba_diff, other_diff)

# Print the mean absolute seniority difference for editors in AB-BA sequences and editors in the rest of the network
result = mean_diff(abba_diff, other_diff)
print(result)
```
